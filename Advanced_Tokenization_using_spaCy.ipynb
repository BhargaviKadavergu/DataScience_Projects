```python
import spacy
import math
from collections import Counter
nlp=spacy.load("en_core_web_sm")
sample_input=["Hello Bhargavi! Welcome to Skill IT. Please visit us at www.skillit.co.in. reach out to support@skillit.com. Thank you for visiting skill IT"]
procfile=[]
# Step 1 : Perform Word Tokenization
print("..........Word Tokenization..........")
for a in sample_input:
  d=nlp(a)
  w=[token.text.lower() for token in d if not token.is_stop and not token.is_punct]
  procfile.append(w)
print(procfile)
tf=[]
for tokens in procfile:
  termfreq=Counter(tokens)
  totalcount=len(tokens)
  for word,count in termfreq.items():
    termfreq[word]=count/totalcount
  tf.append(termfreq)
print(tf)

#Step 2: Perform Sentence Tokenization
print("................Sentence Tokenization..............")

for sent in d.sents:
  print(sent.text)

#Step 3: Print token index positions
print("..........Token Index Positions......")
idf={}
totaldoc=len(procfile)
print("Length of procfile is : ", totaldoc)
allwords=set(word for tokens in procfile for word in tokens)
for word in allwords:
  cont=0
  for doc_tokens in procfile:
    if word in doc_tokens:
      cont+=1
  idf[word]=math.log(totaldoc/cont)
print(idf)

tfidf={}
for word,freq in tf[0].items():
  tfidf[word]=round(freq*idf[word],2)
print(tfidf)

#Step 4: Count total number of tokens
print("....Total Number of tokens")
wordcount=Counter(w)
print(wordcount)
totalwords=len(w)
print("total no of words in the sentence",totalwords)
for word,count in wordcount.items():
  TF=count/totalwords
  print(f"{word},{TF}")

#Step 5: Identify punctuation tokens.
print("...........Punctuation Tokens...........")
for token in d:
    if token.is_punct:
        print(token.text)
```
