```python
from collections import Counter
import spacy
nlp=spacy.load("en_core_web_sm")
text="Hi Bhargavi Welcome!. Please reach to admin team & Billing team % *. Please visit www.skillit.co.in. Thank you"
d=nlp(text)
w=[token.text.lower() for token in d if not token.is_stop and not token.is_punct]
print(w)
wordcount=Counter(w)
print(wordcount)
for sent in d.sents:
  print(sent.text)
```
